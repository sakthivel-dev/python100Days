```python
# Web Scraping with Beautiful Soup and Requests.

# Web scraping involves extracting data from websites. 
# The requests library is used to fetch web pages, and 
# Beautiful Soup is a popular library for parsing and navigating HTML content

import requests
# the flipkart link is under 20k mobiles filter it.
response = requests.get("https://www.flipkart.com/search?q=under%2020k%20mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
html_content = response.content

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find all <a> tags
all_links = soup.find_all("a")

# Use CSS selectors with .select()
selected_elements = soup.select(".class-name")

# Extract attribute value
link_href = all_links[0]["href"]
```

```python
# unwanted words romove
words_to_remove = ["Explore Plus", "Login", "Become a Seller", "Cart","Mobiles & Accessories", "Mobiles", 
                   "Need help?Help me decide", "Home","Add to Compare","044-45614700","Advertise","Gift Cards","Help Center"]
processed_links = set()

for link in all_links:
    link_text = link.text
    
    # Remove specific words from link_text
    for word in words_to_remove:
        link_text = link_text.replace(word, "").strip()
    
    if link_text and link_text not in processed_links:
        processed_links.add(link_text)
        link_url = link["href"]
        print(link_text+"\n")
```
