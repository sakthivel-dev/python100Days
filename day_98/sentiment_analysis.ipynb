```python
from bs4 import BeautifulSoup
import requests
from lxml import html
text_review=[]

# URL of the webpage with reviews
url = "https://www.flipkart.com/machine-learning-using-python/product-reviews/itmb674ebe92d32a?pid=9788126579907&lid=LSTBOK9788126579907FSK5VJ&marketplace=FLIPKART"

# Define the common part of the XPath for reviews
review_base_xpath = '//*[@id="container"]/div/div[3]/div/div/div[2]'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text

    # Parse the HTML content using lxml.html
    tree = html.fromstring(html_content)

    # Iterate through the range of review indices (3 to 12 in your example)
    for review_index in range(3, 13):
        # Construct the complete XPath for the current review
        review_xpath = f'{review_base_xpath}/div[{review_index}]/div/div/div/div[2]'
#         print(review_xpath)
        
        # Find the individual review element using the constructed XPath
        review = tree.xpath(review_xpath)

        # Check if the review was found
        if review:
            review_text = review[0].text_content()
            print(f"Review {review_index}: {review_text}",'/n')
            print("=" * 30)
            text_review.append(review_text)
        else:
            print(f"Review {review_index} not found.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```
