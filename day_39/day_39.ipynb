```python
# A rug plot is a simple and informative plot type commonly used to visualize
#  the distribution of data along a single axis. It consists of small tick marks 
# (usually vertical) that represent individual data points.
#  Rug plots are often used in combination with other visualization techniques,
#  such as histograms or kernel density plots, to provide additional context to 
# the data distribution.
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [3.5, 2.1, 1.0, 2.8, 4.5, 3.1, 2.6, 3.8, 4.1, 2.3]

# Create a rug plot
sns.rugplot(data, height=0.2, color="blue", alpha=0.7)

# Set plot labels and title
plt.xlabel("Data Values")
plt.title("Rug Plot Example")

# Show the plot
plt.show()
```
